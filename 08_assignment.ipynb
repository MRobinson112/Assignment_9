```

# Load the dataset
file_path = '/content/drive/MyDrive/Electric_Vehicle_Population_Data.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()
```

```
# Summary statistics
df.describe()
```

```
# Checking for missing values
df.isnull().sum()

# Visualizing missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```
# Value counts of a categorical column
df['Make'].value_counts()

# Distribution of Electric Vehicle Types
plt.figure(figsize=(12, 6))
df['Electric Vehicle Type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Electric Vehicle Types')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.show()
```

```
#  Filling missing values
df['Model Year'].fillna(df['Model Year'].mode()[0], inplace=True)
```

```
# Scatter Plot
plt.figure(figsize=(20, 6))
plt.scatter(df['Model Year'], df['Electric Range'], color='blue', label='Data Points')
plt.title('Scatter Plot of Electric Range vs. Model Year')
plt.xlabel('Model Year')
plt.ylabel('Electric Range (miles)')
plt.legend(loc='upper right', fontsize='small')
plt.show()

# Bar Plot
plt.figure(figsize=(20, 6))
df['Make'].value_counts().plot(kind='bar', color='green')
plt.title('Bar Plot of Electric Vehicle Makes')
plt.xlabel('Make')
plt.ylabel('Counts')
plt.show()

```

```
# Scatter Plot
plt.figure(figsize=(20, 6))
sns.scatterplot(x='Model Year', y='Electric Range', data=df, color='blue')
plt.title('Scatter Plot of Electric Range vs. Model Year')
plt.xlabel('Model Year')
plt.ylabel('Electric Range (miles)')
plt.legend(loc='upper right', fontsize='small')
plt.show()

# Bar Plot
plt.figure(figsize=(20, 6))
sns.countplot(x='Make', data=df, palette='viridis')
plt.title('Bar Plot of Electric Vehicle Makes')
plt.xlabel('Make')
plt.ylabel('Counts')
plt.xticks(rotation=90)
plt.show()
```

Creating plots in matplotlib required a more detailed set of code to customize some aspects of the plot, such as labels, titles, and legends. where as, seaborn provided an easier way for creating statistical graphics, with less code for customization. For instance, seaborn automatically styles the plots and provides functions to easily visualize complex datasets.
